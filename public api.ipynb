```python
# xmltodict : xml을 dixt 형식으로 변환시켜준다
import requests,xmltodict,json
```

### 웹에서 데이터 받아오기

```python
key = "IEGM3Rx0ZuvXVDI%2BsBth8oY59lmO9loDNpOLFvySEPboEXwHs%2FVJG5k03X5PyQ2nG3kuGdY2aTCIyiv5aw5DOg%3D%3D"
```

```python
url = "http://apis.data.go.kr/6480000/gyeongnamsigundisorder/gyeongnamsigundisorderlist?serviceKey={}&numOfRows=1000&pageNo=1".format(key)
```

```python
# content에 웹에서 가져온 내용을 담는다
content = requests.get(url).content
```

```python
print(content)
```

### xml을 dict로 변환하기

```python
dict = xmltodict.parse(content)
```

### 파이썬 객체를 JSON파일로 저장하는 json.dumps()를 사용
#### 가독성을 높이기 위해  indent=4를 설정

```python
jsonString = json.dumps(dict['response']['body']['items']['item'], indent=4)
```

```python
print(jsonString)
```

###  json.loads()로 Python 객체로 읽기 (역직렬화, 디코딩하기)

```python
jsonObj = json.loads(jsonString)
```

```python
print(jsonObj)
```

```python
print(len(jsonObj))
```

### jsonObj의 item을 하나씩 담는다

```python
for item in jsonObj:
    print(item)
```

### item 중에서 추출하고 싶은 것만 추출한다

```python
for item in jsonObj:
    print(item['signgunm'])
```
